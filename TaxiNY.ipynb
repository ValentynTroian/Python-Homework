```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("yellow_tripdata_2019-01.csv")
```

```python
# Додамо словник із зонами NY

zones_glossary = pd.read_csv("taxi_zone_lookup.csv")
```

```python
# Cередня вартість поїздки:

df['total_amount'].mean()
```

```python
# 10 найдорожчих поїздок

df.nlargest(10,'total_amount')
```

```python
# Звідки в середньому найдорожчі поїздки (топ 10)

MostExpensiveRidesPickup = df.groupby('PULocationID').agg(AverageTotalAmount=('total_amount','mean')).nlargest(10,'AverageTotalAmount')

MostExpensiveRidesPickup.merge(zones_glossary,left_on='PULocationID', right_on='LocationID')
```

```python
# Куди в середньому найдорожчі поїздки (топ 10)

MostExpensiveRidesDropoff = df.groupby('DOLocationID').agg(AverageTotalAmount=('total_amount','mean')).nlargest(10,'AverageTotalAmount')

MostExpensiveRidesDropoff.merge(zones_glossary,left_on='DOLocationID', right_on='LocationID')
```

```python
# 10 найдорожчих напрямків (додамо фільтр, що по напрямку має мати щонайменше 5 поїздок)

directions = df.groupby(['PULocationID','DOLocationID']).agg(AverageTotalAmount=('total_amount', 'mean'), TotalRides=('total_amount', 'count'))
directions.loc[directions['TotalRides'] > 5].nlargest(10,'AverageTotalAmount')
```

```python
# Діапазон цін для 10% найдешевших поїздок
# 0.1 = 10% від всіх рядків

MinTotalAmount = df['total_amount'].min()
MaxTotalAmount = df['total_amount'].sort_values().head(int(len(df)*0.1)).max() 
print(f'Price range at 10% of the cheapest trips is from {MinTotalAmount} to {MaxTotalAmount}')
```

```python
# А якщо забрати поїздки з від'ємним значенням total_amount?

df['total_amount'].loc[df['total_amount'] > 0].sort_values().head(int(len(df)*0.1)).max() 

print(f'Price range at 10% of the cheapest trips is from 0 to {MaxTotalAmount}')
```

```python
df['ParsedPickupHour'] = pd.to_datetime(df['tpep_pickup_datetime'],format='%Y-%m-%d %H:%M:%S.%f').dt.hour
```

```python
# Кількість поїздок по годинах (0 - від 0:00 до 0:59 і т.д.)

df.groupby(['ParsedPickupHour']).agg(Count=('ParsedPickupHour','count'))
```

```python
# Підрахунок к-сті поїздок в залежності від години
# (-1, 5] = з 00:00 до 05:59 - ніч
# (5, 11] = з 06:00 до 11:59 - ранок
# (11, 17] = з 12:00 до 16:59 - день
# (17, 23] = з 17:00 до 23:59 - вечір

df.groupby(pd.cut(df['ParsedPickupHour'],np.arange(-1,24,6),include_lowest=False)).agg(Count=('ParsedPickupHour','count'))
```

```python
# Отже, сумарно зранку і ввечері поїздок більше, ніж вночі і днем. Але днем їх найбільше з усіх 4 періодів
```

```python
# Нехай коротка поїздка - це поїздка до 3 км
```

```python
df['ParsedPickupMonth'] = pd.to_datetime(df['tpep_pickup_datetime'],format='%Y-%m-%d %H:%M:%S.%f').dt.month

df[(df.trip_distance > 0) & (df.trip_distance < 3)].groupby('ParsedPickupMonth').count().max(axis=1).rename('count').reset_index()
```
